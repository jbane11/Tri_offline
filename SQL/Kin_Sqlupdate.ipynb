```python
debug=1
RLdir = '/home/jbane/tritium/Tri_offline/SQL/Runlist/'
```

```python

def run_search(run) :
    
    locations=[]
    filenames=[]
    for files in glob.glob('{}*.dat'.format(RLdir) ):
        f = open( files, 'r' )
        file_contents = f.read()
        if str(run) in file_contents:
            if "9{}".format(run) in file_contents:
                continue 
            else :
                locations.append(f.name)
                if debug ==1:
                    print (f.name)
                
                            
                            

        f.close()
    if len(locations) > 0 :
        for j in locations :
            filenames.append(j.replace(RLdir,""))

        if debug ==1 :
            print(filenames)

        if len(filenames) == 1 :
            pos = filenames[0].find("_")
            tgt = filenames[0][:pos]
            
            substr =filenames[0].replace(tgt,"")
            substr =substr.replace(".dat","")
            pos1 = substr.find("kin")
            kin = substr[pos1+3:]

            if "p" in tgt:
                kin=kin+"_pos"
                tgt=tgt.replace("p","")
            
            return tgt,kin
        else :
            print("More then one choice")
            return "uk","uk"
    else :
        return "uk","uk"
        
        
    
```

```python
firstr=1209
lastr=1400
badruns=[]
for run in range(firstr,lastr) :
    tgt,kin =run_search(run)
    if tgt == 'uk':
        badruns.append(run)
print(badruns)
```
